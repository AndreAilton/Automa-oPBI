```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from time import sleep
from pathlib import Path
from selenium.webdriver import ActionChains
import os
import time

def VerificarDiretorio():
    caminho = Path(os.path.expanduser('~')) / 'Documents'
    if Path(caminho / 'AutomacaoS').exists():

        if Path(caminho / 'AutomacaoS' / 'Imprimir').exists():
            return Path(caminho / 'AutomacaoS' / 'Imprimir')
        else:
            Path(caminho / 'AutomacaoS' / 'Imprimir').mkdir()
            return Path(caminho / 'AutomacaoS' / 'Imprimir')
    else:
        Path(caminho / 'AutomacaoS').mkdir()
        Path(caminho / 'AutomacaoS' / 'Imprimir').mkdir()
        return Path(caminho / 'AutomacaoS' / 'Imprimir')


def BaixarPowerBi(login, senha, links):
    tempo_inicial = time.time()
    Caminho = str(VerificarDiretorio())
    loginw = 'https://login.microsoftonline.com/'

    options = webdriver.ChromeOptions()
    #options.add_argument("--headless=new")
    options.add_experimental_option("prefs", {
        "download.default_directory": f"{Caminho}",
        "download.prompt_for_download": False,  # Desativa a caixa de diálogo de confirmação de download
        "download.directory_upgrade": True,
        "safebrowsing.enabled": True})

    servico = Service(ChromeDriverManager().install())
    navegador = webdriver.Chrome(options=options,service=servico)
    print('Inicializando Navegador')
    navegador.get(loginw)
    sleep(10)
    try:

        print('Fazendo Login na Microsoft')
        navegador.find_element(By.XPATH, '//*[@id="i0116"]').send_keys(login)
        navegador.find_element(By.XPATH, '//*[@id="idSIButton9"]').click()
        sleep(5)
        navegador.find_element(By.XPATH, '//*[@id="i0118"]').send_keys(senha)
        navegador.find_element(By.XPATH, '//*[@id="idSIButton9"]').click()
        sleep(5)
        navegador.find_element(By.XPATH, '//*[@id="idSIButton9"]').click()
        sleep(5)

        print('Abrindo Indicadores Aguarde')
        for link in links:
            navegador.execute_script("window.open('', '_blank');")
            navegador.switch_to.window(navegador.window_handles[-1])
            navegador.get(link)
            sleep(20)
            navegador.find_element(By.XPATH, '//button[(@id = "exportMenuBtn")]').click()
            sleep(3)
            navegador.find_element(By.XPATH, '//button[@data-testid="export-to-pdf-btn"]').click()
            sleep(5)
            navegador.find_element(By.XPATH, '//button[@id="okButton"]').click()
            sleep(5)

            print('Baixando Indicadores')
            for c in range(40):
                print('l',end='')
                sleep(1)

        print('')
        print(f'Tempo Gasto total: {round((time.time() - tempo_inicial)/ 60,2)} minutos')
    except:
        print('Verifique Senha e links')


def Atualizarindicador(login,senha,workspace):
    tempo_inicial = time.time()
    loginw = 'https://login.microsoftonline.com/'
    workspace= workspace
    options = webdriver.ChromeOptions()
    #options.add_argument("--headless=new")
    servico = Service(ChromeDriverManager().install())
    navegador = webdriver.Chrome(options=options, service=servico)
    print('Inicializando Navegador')
    sleep(5)
    try:
        navegador.get('https://www.google.com/')
        sleep(5)
        navegador.get(loginw)
        sleep(5)
        print('Fazendo Login na Microsoft')
        navegador.find_element(By.XPATH, '//*[@id="i0116"]').send_keys(login)
        navegador.find_element(By.XPATH, '//*[@id="idSIButton9"]').click()
        sleep(3)
        navegador.find_element(By.XPATH, '//*[@id="i0118"]').send_keys(senha)
        navegador.find_element(By.XPATH, '//*[@id="idSIButton9"]').click()
        sleep(3)
        navegador.find_element(By.XPATH, '//*[@id="idSIButton9"]').click()
        navegador.get(workspace)
        sleep(6)

        print('Filtrando Conjuntos de Dados!')
        navegador.find_element(By.XPATH,
                               "//button[@class='tri-filter-menu-trigger-button tri-flex tri-justify-center tri-items-center tri-text-sm tri-border tri-rounded tri-border-solid tri-bg-gray-white tri-border-gray-82 tri-h-32px tri-text-gray-180 tri-box-border tri-filter-menu-state']").click()
        sleep(3)
        navegador.find_element(By.XPATH, "//div[@id='1'][@class='tri-flex tri-items-center tri-multiselect-list-item-label tri-w-full tri-h-32px tri-px-2']").click()

        print('Atualizando Indicadores')
        actions = ActionChains(navegador)
        lista =  navegador.find_elements(By.XPATH, "//mat-icon[@data-mat-icon-name='pbi-glyph-refresh']")
        print('Aguarde: ',end='')

        sleep(4)
        for valor in range(len(lista)):
            elemento = navegador.find_elements(By.XPATH, "//mat-icon[@data-mat-icon-name='pbi-glyph-refresh']")[valor]
            sleep(2)
            actions.move_to_element(elemento).perform()
            sleep(2)
            elemento.click()
            sleep(5)
            print('l',end='')

        print('')
        print(f'Tempo Gasto total: {round((time.time() - tempo_inicial)/ 60,2)} minutos')
    except:
        print('Verifique Senha e Workspace do Powerbi')
```
